#### Importing libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import statistics as stats
```

#### Reading datasets

```python
steps=pd.read_csv("steps.csv", sep=';')
steps["mean"] = steps.iloc[:,:].mean(axis=1)
steps.head(15)
```

```python
survey=pd.read_csv("survey.csv")
survey.info()
survey.head()
```

```python
df1 = pd.merge(steps, survey, on='id')
df1.head(15)
```

```python
sns.histplot(data=df1, x="mean", kde=True)
```

De histogram geeft een rechter helling aan. De figuur maakt duidelijk dat de meeste observeringen gemiddeld minder dan 25000 stappen zetten. 

```python
df1['mean'].median(), df1['mean'].mean()
```

Gemiddeld worden er 6811 stappen gezet. Het middelste getal in de verdeling is: 6078,955

```python
mean = df1['mean']
stats.variance(mean)
```

De variance is erg groot. Dit laat zien dat de getallen erg uit elkaar liggen. Dit laat zien dat er erg veel outliers zijn.
