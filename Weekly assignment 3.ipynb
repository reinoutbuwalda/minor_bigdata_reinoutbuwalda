<h3>Wat is lineaire regressie?</h3>

Als eerst wordt er bij lineaire regressie gebruik gemaakt van de correlatie techniek om een verband te constateren. 
Bijvoorbeeld de samenhang tussen de temperatuur en het aantal positieve corona testen. Bij lineaire regressie
ga je net een stapje verder. Dan probeer je namelijk de waardes van Y te voorspellen met een veranderende variabele
X. Aspecten die komen kijken bij lineaire regressie zijn: de regressie lijn, de b coefficient (hellling) en de a 
coëfficient die het nul punt bepaalt.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import math
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
DataFrame= pd.read_csv("DataFrame_week3.csv")
#DataFrame.rename(columns={'Unnamed: 0': 'poep'})
DataFrame.columns
DataFrame.rename(columns = {'Unnamed: 0': 'Unnamed'}, inplace=True)
DataFrame.drop('Unnamed', axis=1, inplace=True)
DataFrame.head(15)
```

```python
DataFrame['Sex_male']= DataFrame.gender.map({'Female':0, 'Male':1})
DataFrame.head(25)
```

```python
DataFrame.city.value_counts(), DataFrame.education_1.value_counts(), DataFrame.age.value_counts()
```

```python
city_dumies = pd.get_dummies(DataFrame.city, prefix='city')
education_dummies = pd.get_dummies(DataFrame.education_1, prefix='education')
age_dummies = pd.get_dummies(DataFrame.age, prefix='age')
```

```python
df = pd.concat([DataFrame, city_dumies, education_dummies, age_dummies], axis=1)
df.head(15)
```

Drop all the variables we don't want to use

```python
df1= df.drop(columns=['id','age','height_div','weight','gender','education_1','city'], axis=1)
df1.head(15)
```

## Fitting the model

We will split our data, train our model and then test it. 

6. Let's first create the *X* and *y* variables.

```python

y = df1['mean']
X = df1.loc[:,['bmi','Sex_male','city_Bordeaux', 'city_Lille', 'city_Lyon',
       'city_Montpellier', 'education_Bachelor', 'education_High school',
       'education_Master or doctorate', 'education_No degree',
       'education_Primary', 'education_Vocational', 'age_25-34', 'age_35-44',
       'age_45-54', 'age_55-64', 'age_65+', 'age_<25']]
X.head()
```

Split the data in 70/30, 70% training set and 30% test set.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)
X_train.head(15)
```

Now let's create the model and train it. Also calculate the intercept and slopes

```python
lm = LinearRegression()
model = lm.fit(X_train, y_train)
```
